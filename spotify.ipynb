```python
def get_top_tracks(artist_name):
    try:
        # Search for the artist
        results = sp.search(q=f"artist:{artist_name}", type="artist", limit=1)
        if not results['artists']['items']:
            return f"Artist '{artist_name}' not found on Spotify."

        artist = results['artists']['items'][0]
        artist_id = artist['id']

        # Get the artist's top tracks
        top_tracks = sp.artist_top_tracks(artist_id) # returns top 10 tracks
        if not top_tracks['tracks']:
            return f"No top tracks found for artist '{artist_name}'."

        # Display the top 5 tracks
        print(f"All tracks for {artist['name']}:\n")
        for i, track in enumerate(top_tracks['tracks'], 1):
             print(f"{i}. {track['name']} (Popularity: {track['popularity']})")

    except Exception as e:
        return f"An error occurred: {e}"
```

```python
# Input the artist name
artist_name_input = input("Enter the artist's name: ")
get_top_tracks(artist_name_input)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

# Spotify API credentials
CLIENT_ID = "your_spotify_client_id"  # Replace with your Client ID
CLIENT_SECRET = "your_spotify_client_secret"  # Replace with your Client Secret

def get_artist_tracks_with_details(artist_name):
    try:
        # Step 1: Search for the artist
        results = sp.search(q=f"artist:{artist_name}", type="artist", limit=1)
        if not results['artists']['items']:
            return f"Artist '{artist_name}' not found on Spotify."

        artist = results['artists']['items'][0]
        artist_id = artist['id']
        print(f"Found artist: {artist['name']} (ID: {artist_id})")

        # Step 2: Get all albums by the artist
        albums = sp.artist_albums(artist_id, album_type='album,single', limit=50)
        album_details = [
            {"id": album['id'], "name": album['name'], "release_date": album['release_date']}
            for album in albums['items']
        ]

        # Remove duplicate albums by name
        unique_albums = {album['name']: album for album in album_details}.values()
        print(f"Found {len(unique_albums)} unique albums.")

        # Step 3: Get all tracks with details from each album
        all_tracks = []
        for album in unique_albums:
            album_id = album['id']
            album_name = album['name']
            album_year = album['release_date'].split('-')[0]  # Extract the year from release date

            album_tracks = sp.album_tracks(album_id)
            for track in album_tracks['items']:
                # Fetch track details to get the popularity score
                track_details = sp.track(track['id'])
                all_tracks.append({
                    "track_name": track['name'],
                    "album_name": album_name,
                    "release_year": album_year,
                    "popularity": track_details['popularity']
                })

        # Remove duplicates by track name
        unique_tracks = {track['track_name']: track for track in all_tracks}.values()

        # Step 4: Convert to Pandas DataFrame and sort by popularity
        df = pd.DataFrame(unique_tracks)
        df.sort_values(by="popularity", ascending=False, inplace=True)

        print(f"Collected {len(df)} unique tracks for {artist_name}.")
        return df

    except Exception as e:
        return f"An error occurred: {e}"
```

```python
# Input the artist name
artist_name_input = input("Enter the artist's name: ")
result = get_artist_tracks_with_details(artist_name_input)
```

```python
# If a DataFrame is returned, save it to a CSV file and display it
if isinstance(result, pd.DataFrame):
    print("\nTop tracks sorted by popularity:")
    print(result.head())  # Display the top tracks
    result.to_csv(f"{artist_name_input}_tracks_with_details.csv", index=False)  # Save to CSV
    print(f"\nTrack data saved to {artist_name_input}_tracks_with_details.csv")
else:
    print(result)
```
